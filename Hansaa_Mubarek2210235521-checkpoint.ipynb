## Giriş
Bu proje kapsamında, Hastalık Belirtileri ve Hasta Profili Veri Kümesinin derinlemesine bir analizini yapmak için makine öğrenimi (ML) bilgimi kullandım. Veri seti, pozitif veya negatif olarak etiketlenen sonuçlarla birlikte hastalık semptomlarıyla ilgili bilgileri içerir. Ek olarak, hastaların yaş ve cinsiyet gibi demografik ayrıntılarının yanı sıra tıbbi geçmişleri, özellikle kronik hastalıkları içerir.

Analizin birincil amacı, tanımlanamayan bir salgının test sonucunu temsil eden "sonuç değişkenini" incelemektir. Salgın ateş, öksürük, yorgunluk ve nefes almada zorluk gibi yaygın semptomlarla karakterizedir. Amaç, test sonucunun (pozitif veya negatif olarak kategorize edilir) hem hastanın tıbbi geçmişi (Hastalık sütununda yakalanır) hem de demografik özellikleri (ör. kan basıncı veya yaş) ile nasıl ilişkili olduğuna dair içgörüler elde etmektir.

## Sütun:
Hastalık: Hastalığın veya tıbbi durumun adı. Ateş: Hastanın ateşi olup olmadığını gösterir (Evet/Hayır). Öksürük: Hastanın öksürüğü olup olmadığını gösterir (Evet/Hayır). Yorgunluk: Hastanın yorgunluk yaşayıp yaşamadığını gösterir (Evet/Hayır). Nefes Almada Zorluk: Hastanın nefes almakta güçlük çekip çekmediğini gösterir (Evet/Hayır). Yaş: Hastanın yıl cinsinden yaşı. Cinsiyet: Hastanın cinsiyeti (Erkek/Kadın). Kan Basıncı: Hastanın kan basıncı seviyesi (Normal/Yüksek). Kolesterol Seviyesi: Hastanın kolesterol seviyesi (Normal/Yüksek). Sonuç Değişkeni: Spesifik hastalık için tanı veya değerlendirmenin sonucunu gösteren sonuç değişkeni (Pozitif/Negatif).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error



x=df.iloc[:,:8]
y=df.iloc[:,:-1]
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("mlr:","r2:",r2_score(y,mlr_pred),"mae:",mean_absolute_error(y,mlr_pred))
print("dt:","r2:",r2_score(y,dt_pred),"mae:",mean_absolute_error(y,dt_pred))
print("rf:","r2:",r2_score(y,rf_pred),"mae:",mean_absolute_error(y,rf_pred))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)


mlr=LinearRegression()
mlr.fit(x_train,y_train)
mlr_pred=mlr.predict(x_test)

dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt_pred=dt.predict(x_test)

rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
rf_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("mlr:","r2:",r2_score(y_test,mlr_pred),"mae:",mean_absolute_error(y_test,mlr_pred))
print("dt:","r2:",r2_score(y_test,dt_pred),"mae:",mean_absolute_error(y_test,dt_pred))
print("rf:","r2:",r2_score(y_test,rf_pred),"mae:",mean_absolute_error(y_test,rf_pred))
```
